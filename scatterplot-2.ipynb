```r
# Fig2A
# sensitivity 
x <- c(0,0.1,0.25,0.5,0.75,1)
Firefly <- c(0.697, 0.720, 0.776, 0.836, 0.898, 0.993)
Kafka <- c(0.387, 0.442, 0.557, 0.719, 0.860, 0.970)
Bronya <- c(0.392, 0.420, 0.525, 0.703, 0.843, 0.985)
SilverWolf <- c(0, 0.313, 0.515, 0.680, 0.828, 0.937)
BlackSwan <- c(0.319, 0.333, 0.481, 0.643, 0.807, 0.931)
StarRail <- data.frame(x=x, Kafka=Kafka, Bronya=Bronya, SilverWolf=SilverWolf, BlackSwan=BlackSwan, Firefly=Firefly)
StarRail$x <- factor(StarRail$x, levels = x) 

df <- reshape2::melt(StarRail,id.vars="x")
df <- df[-which(df$x==0 & df$variable=="SilverWolf"),]
df$x <- as.numeric(as.character(df$x))
head(df)
```

```r
library(ggplot2)
ggplot(data = df, aes(x = x, y = value, colour = variable, group = variable)) + 
  geom_line(linewidth = 0.5) +
  geom_point(size = 2, shape = 21, fill = "white", stroke = 1) +
  scale_y_continuous(breaks = c(0.25, 0.5, 0.75, 1.0), limits = c(0.25, 1.0), labels = c("0.25", "0.5", "0.75", "1")) + 
  theme_classic() +
  theme(
        plot.margin = unit(c(1, 1, 1, 1), "cm"),
        axis.ticks.length = unit(0.2,'cm'),
        axis.ticks = element_line(color = "black",linewidth = 0.5),
        axis.line = element_line(color = "black",linewidth = 0.5),
        axis.text.x = element_text(color = "black"),
        axis.text.y = element_text(color = "black"),
        panel.border = element_blank(),
        panel.grid = element_blank(),
        legend.title = element_blank(),
        legend.margin = margin(0, 0, 0, 0),
        legend.text = element_text(size = 18, color = "black"),
        legend.position = "bottom",
        legend.key.width = unit(14, "mm"),
        legend.key.spacing.y = unit(0.3, "cm"),
        text = element_text(size = 20)) +
  guides(colour = guide_legend(nrow = 2, byrow = TRUE)) +
  scale_color_manual(values=c("#5b95d0", "#f39c43","#1bb174","#999999","#ee7d7d")) +
  labs(x="Star Rail", y="Character value")
```
