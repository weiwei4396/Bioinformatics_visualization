****1.ggradar****

```r
Firefly <- c(0.559, 0.588, 0.662, 0.75, 0.838, 1)
Kafka <- c(0.265, 0.309, 0.426, 0.603, 0.765, 0.956)
Bronya <- c(0.294, 0.309, 0.382, 0.574, 0.75, 0.971)
SilverWolf <- c(0, 0.147, 0.353, 0.559, 0.765, 0.985)
BlackSwan <- c(0.265, 0.25, 0.382, 0.544, 0.706, 0.897)

StarRail <- data.frame(Kafka = Kafka , Bronya = Bronya, SilverWolf = SilverWolf, BlackSwan = BlackSwan, Firefly = Firefly)
StarRail <- data.frame(t(StarRail))
colnames(StarRail) <- c("0%","10%","25%","50%","75%","100%")
StarRail$group <- rownames(StarRail)
StarRail <- StarRail[,c("group", "0%","10%","25%","50%","75%","100%")]
StarRail$group <- factor(StarRail$group, levels = c("Bronya", "Firefly", "Kafka", "SilverWolf", "BlackSwan"))
StarRail
```

```r
# 缺点是只能三个背景值;
library(ggradar)
library(ggplot2)
ggradar(
  StarRail, 
  values.radar = c("0", "0.5", "1"),# 最小，平均和最大网格线显示的值
  grid.min = 0, # 绘制最小网格线的值
  grid.mid = 0.5, # 绘制平均网格线的值
  grid.max = 1, # 绘制最大网格线的值
  grid.line.width = 0.5,
  gridline.min.colour = "grey",
  gridline.mid.colour = "grey",
  gridline.max.colour = "grey",
  gridline.min.linetype = "longdash",
  gridline.mid.linetype = "longdash",
  gridline.max.linetype = "longdash",
  grid.label.size = 3,
  # 0 0.5 1;
  label.gridline.min = FALSE,
  label.gridline.mid = FALSE,
  label.gridline.max = FALSE,
  # 0 0.5 1 的大小;
  axis.label.size = 9, # 100% 75% 的大小;
  group.line.width = 0.6, 
  group.point.size = 0,
  group.colours = c("#f39c43", "#ee7d7d", "#5b95d0", "#1bb174", "#999999"),
  # Background and grid lines
  background.circle.colour = "white",
  background.circle.transparency = 0.3,
  legend.position = "bottom",
  legend.text.size = 20,
  fill = FALSE,
  fill.alpha = 0.2,
  plot.title = "Characters") +
  theme(plot.title = element_text(hjust = 0.5)) +
  guides(colour = guide_legend(nrow = 2, byrow = TRUE))
```
